```python
from sklearn.model_selection import train_test_split
```

```python
data_home.columns
```

```python
data_home=data_home.drop(['Row Labels'],axis=1, inplace=True)
```

```python
x=data_home.drop(['Sum of CSUSHPISA','Row Labels'], axis=1)
y=data_home['Sum of CSUSHPISA']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
coff_data_Home=pd.DataFrame(lm.coef_,x.columns,columns=['Coefficient'])
coff_data_Home
```

```python
pred=lm.predict(x_test)
```

```python
plt.scatter(y_test,pred)
plt.show()
```

```python
sns.distplot((y_test-pred),bins=50)
```
